## Polinomio de Newton

```python
import numpy as np
import matplotlib.pyplot as plt

def difDiv(X,Y):
    '''
    X = arreglo de nodos
    Y = arreglo de abcisas
    n = número de diferencias
    '''
    n = X.size    
    for i in range(n):
        for j in range(n-1, i, -1):
            Y[j]=(Y[j]-Y[j-1])/(X[j]-X[j-i-1])
    return Y

def polinomio(X,Y,a):
    suma = 0
    e = 1
    n = X.size
    for i in range(n):
        suma = suma + e*Y[i]
        e = e*(a-X[i])
    return suma

def horner(X,Y,a): 
    n = Y.size-1
    b = Y[-1] ## b = Y[n=4]
    for k in range(n-1,-1,-1): ## de 3 a 0  ## o de -2 a -5
        b = Y[k] + b*(a-X[k])
    return b
    
#IMPLEMENTACIÓN

x = np.array([100.,120,150,180,210])
y = np.array([124.1,181.9,285.5,457.4,551.6])
plt.plot(x,y,'o')
print(difDiv(x,y))
print(polinomio(x,y,160))
print(horner(x,y,160))
print(polinomio(x,y,220))
plt.show()    
```

```python
import numpy as np
import scipy.interpolate as ip
import matplotlib.pyplot as plt

X = np.array([100.,120,150,180,210,235])
Y = np.array([124.1,181.9,285.5,457.4,551.6,679.2])

f = ip.CubicSpline(X,Y)
x = np.linspace(X[0],X[-1],200)

plt.plot(X,Y,'o',label="datos")
plt.plot(x,f(x),label="interpolación")
plt.title("Cubic Spline")
plt.legend()
plt.show()

print(f(160))
print(f(220))
```
