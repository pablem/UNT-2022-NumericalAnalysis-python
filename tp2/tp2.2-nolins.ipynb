    b) Se proponen las siguientes funciones de iteraci√≥n:

$ùëî_1(ùë•) = \sqrt{2ùë• + 3} $

$ùëî_2(ùë•) = \frac{ùë•^2‚àí3}{2}$

$ùëî_3(ùë•) = \frac{3}{ùë•‚àí2}$

Implemente el m√©todo de punto fijo para aproximar las ra√≠ces de la funci√≥n f(x) con cada una de estas funciones, calculando el error en cada iteraci√≥n.

Utilizando los siguientes valores iniciales:

i) ‚Äì 4 ii ) -0.99 iii) 3.5 iv) 2 v) 0

- ¬øC√≥mo influye la elecci√≥n de la funci√≥n de iteraci√≥n? Y de los puntos iniciales?
- Grafique en un mismo eje cartesiano cada funci√≥n g(x), f(x) y la primera bisectriz. ¬øQu√© puede observar?


```python
import numpy as np

x = np.linspace(-4, 4, 100)

def f(x):
    return x**2 - 2*x -3 ##ra√≠ces en x=-1, x=3

def g1(x):
    return np.sqrt(2*x+3)

def g2(x):
    return (x**2-3)/2

def g3(x):
    return 3/(x-2)

### M√©todo de punto fijo:
def puntoFijo(a, itmax, eps, g):
    
    it = 0
    error = eps + 1
    
    print("\nIteraciones: ", end="")
    
    while it <= itmax and error > eps:
        it = it + 1
        b = g(a)
        error = np.abs(b-a)
        print('{:<.2e}'.format(error), '  ', end="")
        a = b

    if it > itmax:
        print("\nNo converge")
    else: 
        print("\nLa ra√≠z es:  ", '{:.6f}'.format(b), "\tError: ", error, " N¬∞ de iteraciones: ", it)
        
### implementaci√≥n...
itmax = 20
eps   = 1E-4

print("\n--- g1 ---")
print("\nx = -4")
# puntoFijo(-4, itmax, eps, g1)
print("Ra√≠z de un n√∫mero negativo.")
print("\nx = -0.99")
puntoFijo(-0.99, itmax, eps, g1)
print("\nx = 3.5")
puntoFijo(3.5, itmax, eps, g1)
print("\nx = 2")
puntoFijo(2, itmax, eps, g1)
print("\nx = 0")
puntoFijo(0, itmax, eps, g1)

print("\n\n--- g2 ---")
print("\nx = -4")
# puntoFijo(-4, itmax, eps, g2) ##OverflowError: (34, 'Result too large')
print("No converge, a-b increment√≥ exponencialmente")
print("\nx = -0.99")
# puntoFijo(-0.99, itmax, eps, g2)
print("No converge, n√∫mero de iteraciones agotado.")
print("\nx = 3.5")
# puntoFijo(3.5, itmax, eps, g2) ##OverflowError: (34, 'Result too large')
print("No converge, a-b increment√≥ exponencialmente.")
print("\nx = 2")
# puntoFijo(2, itmax, eps, g2)
print("No converge, n√∫mero de iteraciones agotado.")
print("\nx = 0")
# puntoFijo(0, itmax, eps, g2)
print("No converge, n√∫mero de iteraciones agotado.")

print("\n\n--- g3 ---")
print("\nx = -4")
puntoFijo(-4, itmax, eps, g3)
print("\nx = -0.99")
puntoFijo(-0.99, itmax, eps, g3)
print("\nx = 3.5")
# puntoFijo(3.5, itmax, eps, g3) ##ZeroDivisionError: float division by zero
print("Divisi√≥n por cero en la segunda iteraci√≥n 3/1.5=2 - 2")
print("\nx = 2")
# puntoFijo(2, itmax, eps, g3) ##ZeroDivisionError: division by zero
print("Divisi√≥n por cero en la primera iteraci√≥n 2 - 2")
print("\nx = 0")
puntoFijo(0, itmax, eps, g3)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 100)

def g3(x):
    return 1/(x+1)

### M√©todo de punto fijo:
def steffensen(a, itmax, eps, g):
    
    it = 0
    error = eps + 1
    
    print("\nIteraciones: ", end="")
    
    while it <= itmax and error > eps:
        it = it + 1
        b = g(a)
        c = g(b)
        error = np.abs(b-a)
        print('{:<.2e}'.format(error), '  ', end="")
        a = a - (a-b)**2 / (c - 2*b + a)

    if it > itmax:
        print("\n\nNo converge")
    else: 
        print("\n\nLa ra√≠z es: ", '{:.6f}'.format(b), "\tError: ", error, "\tN¬∞ de iteraciones: ", it)

### implementaci√≥n...
itmax = 20
eps   = 0.001

print("\nx0 = 1")
steffensen(1, itmax, eps, g3)
```
