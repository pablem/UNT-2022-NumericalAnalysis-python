```python
import numpy as np

def jacobi(x0,a,b,eps,itmax):
    '''
    x0 = punto inicial
    a = matriz coeficientes
    b = vector terminos independientes
    '''
    error = eps+1
    it = 0
    n = x0.size
    x1 = np.zeros(n)
    while it <= itmax and error > eps:
        it = it + 1
        for i in range(n):
            suma = 0
            for j in range(i):
                suma = suma + a[i,j]*x0[j]
            for j in range(i+1,n):
                suma = suma + a[i,j]*x0[j]
            x1[i] = (b[i]-suma)/a[i,i] 
        error = (abs(x1-x0)).max()
        x0 = np.copy(x1)
    print("\n\nLa solución es: ", x1)


##IMPLEMENTACIÓN

eps = 1E-3
it = 20 
a = np.array([[2, -1, 0], [1, 6, -2], [4, -3, 8]])
b = np.array([2, -4, 5])
x = np.array([0, 0, 1])

jacobi(x,a,b,eps,it)
        
```

```python
import numpy as np

def sor(x0,a,b,eps,itmax,w):
    '''
    x0 = punto inicial
    a = matriz coeficientes
    b = vector terminos independientes
    w = coeficiente de ponderación
    '''
    error = eps+1
    it = 0
    n = x0.size
    x1 = np.zeros(n)
    while it <= itmax and error > eps:
        it = it + 1
        for i in range(n):
            suma = 0
            for j in range(i):
                suma = suma + a[i,j]*x1[j]
            for j in range(i+1,n):
                suma = suma + a[i,j]*x0[j]
            x1[i] = (1-w)*x0[i] + w*(b[i]-suma)/a[i,i] 
        error = (abs(x1-x0)).max()
        x0 = np.copy(x1)
    print("\nLa solución es: ", x1, "\tnro iteraciones:", it)


##IMPLEMENTACIÓN

eps = 1E-3
it = 20 
a = np.array([[2, -1, 0], [1, 6, -2], [4, -3, 8]])
b = np.array([2, -4, 5])
x = np.array([0, 0, 1])

for k in np.arange(0.1,2,0.1):
    sor(x,a,b,eps,it,k)
```

```python
import numpy as np
import scipy.special as sps

def gaussLegendre(f,a,b,n):
    nodos, coef = sps.roots_legendre(n)
    I = 0
    for i in range(n):
        I = I + coef[i]*f((b-a)/2*nodos[i]+(b+a)/2)
    I = I*(b-a)/2
    return I

##IMPLE

def f(x):
    return np.exp(x)*np.sin(x)/(1+x**2)

print(gaussLegendre(f,0,2,3))
```

```python
import numpy as np

def simpsonCompuesta(f,a,b,nPtos):
    n = nPtos - 1
    h = (b-a)/n
    par = 0
    impar = 0
    for i in range(1,n):
        x = a+i*h
        if i%2!=0:
            impar = impar + f(x)
        else:
            par = par + f(x)
    return h/3*(f(a)+4*impar+2*par+f(b))
```
